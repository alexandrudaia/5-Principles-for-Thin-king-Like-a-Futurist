```
import matplotlib.pyplot as plt
import keras_ocr
```

```
pipeline = keras_ocr.pipeline.Pipeline()
```

```
 images = [
    keras_ocr.tools.read(img) for img in ['/content/download.png']
]
```

```
 prediction_groups = pipeline.recognize(images)
```

```
prediction = pipeline.recognize(images)
```

```
predicted_image = prediction_groups[0]
for text, box in predicted_image:
    print(text)
```

```
detector = keras_ocr.detection.Detector()
recognizer = keras_ocr.recognition.Recognizer()

pipeline = keras_ocr.pipeline.Pipeline(detector=detector,recognizer=recognizer)
prediction = pipeline.recognize(images)
```

```
prediction
```

```
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))

keras_ocr.tools.drawAnnotations(image=images[0], predictions=prediction_groups[0], ax=axs)
```
